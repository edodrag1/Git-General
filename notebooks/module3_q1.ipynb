```python
class Account:

    def __init__(self, prev_used, limit):

        # Initialise instance variables
        self.prev_used = prev_used
        self.limit = limit

    def check_data_usage(self, data_used):

        # Calculate the new total usage
        total_usage = data_used + self.prev_used

        # Compare total usage against the limit
        if total_usage > self.limit:
            print("Data limit exceeded")
        else:
            print("Data usage is within the limit")

        # Assigning the new total_usage to prev_used
        self.prev_used = total_usage
```

# Instantiate the Account object
my_acct = Account(prev_used=9.4, limit=10)

```python
# Execute the usage check method
my_acct.check_data_usage(0.7)
```
